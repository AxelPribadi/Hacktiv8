## Line Plot Comics Category

```python
gg_comics = google[google["Category"] == "COMICS"]

print(gg_comics.shape)
gg_comics.head(3)
```

```python
gg_comics.set_index("Rating").sort_values("Rating", ascending = False)
```

```python
gg = gg_comics[["Rating"]]
gg.sort_values("Rating", ascending = False)
```

```python
count = np.unique(gg.Rating, return_counts = True)
```

```python
rate = pd.DataFrame(count).T
rate = rate.rename(columns = {0: "Rating", 1: "Frequency"})
```

```python
rate
```

```python
rate.plot("Rating","Frequency", kind = "line")
plt.title("Rating Count in Comics Category")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```
