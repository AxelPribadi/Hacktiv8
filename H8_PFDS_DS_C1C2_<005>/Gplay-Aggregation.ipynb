```python
import pandas as pd
import numpy as np
```

```python
# Open csv file
df = pd.read_csv("googleplaystore.csv")
df.head()
```

```python
df.shape
```

```python
tot = df.isnull().sum().sort_values(ascending = False)
pd.concat([tot], axis = 1, keys = ["Total"])
```

```python
df.dropna(inplace = True)
df.shape
```

```python
print(len(df["Category"].unique()),"Categories", "\n")
print(df["Category"].unique())
```

```python
print(len(df["Genres"].unique()),"Genres", "\n")
print(df["Genres"].unique())
```

## Find Most Expensive Apps

```python
df["Type"].unique()
```

```python
df_paid = df[df["Type"] == "Paid"]
df_paid.head()
```

```python
df_paid.Price = df_paid.Price.apply(lambda x: x.replace('$',''))
df_paid["Price"] = df_paid["Price"].apply(lambda x: float(x))
```

```python
df_paid["Price"].max()
```

```python
df_paid.sort_values("Price", ascending = False).head(10)
```

## Find Least Expensive Apps

```python
df_paid["Price"].min()
```

```python
df_paid.sort_values("Price", ascending = True).head(10)
```

## Count Number of Apps in each Category

```python
df_cat = df.Category.value_counts().reset_index().rename(columns = {"index": "Category", "Category": "Count"})
df_cat
```

## Find Most Downloaded Apps

```python
df.Installs.replace("[\+,]","", regex = True, inplace = True)
```

```python
df.sort_values("Installs", ascending = False).head(10)
```

## Calculate Average Rating in each Category

```python
df.groupby("Category").mean()
```

## Count Number of Apps in each Genre

```python
df_genre = df.Genres.value_counts().reset_index().rename(columns = {"index": "Genres", "Genres": "Count"})
df_genre
```

## Group App by Genre, Find Mean, Median & Mode Rating of each Genre

```python
df.groupby("Genres").mean().sort_values("Rating",ascending = False)
```

```python
df.groupby("Genres").median().sort_values("Rating", ascending = False)
```

```python
df.groupby("Genres").agg(pd.Series.mode)
```

```python
df.groupby("Genres").describe()
```
